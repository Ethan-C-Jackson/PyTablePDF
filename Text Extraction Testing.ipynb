```python
import fitz

def extract_text_with_coordinates(pdf_path):
    doc = fitz.open(pdf_path)
    data = []
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        blocks = page.get_text("blocks")
        for block in blocks:
            x0, y0, x1, y1, text = block[:5]
            data.append((x0, y0, x1, y1, text))
    return data

pdf_path = "ast_sci_data_tables_sample.pdf"
data = extract_text_with_coordinates(pdf_path)
```

```python
from sklearn.cluster import KMeans
import numpy as np

def cluster_text_blocks(data, n_clusters):
    coordinates = np.array([(x0, y0) for x0, y0, x1, y1, text in data])
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(coordinates)
    clusters = kmeans.predict(coordinates)
    clustered_data = [[] for _ in range(n_clusters)]
    for i, cluster in enumerate(clusters):
        clustered_data[cluster].append(data[i])
    return clustered_data

n_clusters = 3  # Adjust based on the expected number of rows/columns
clustered_data = cluster_text_blocks(data, n_clusters)
```

```python
def sort_and_align_blocks(clustered_data):
    sorted_data = []
    for cluster in clustered_data:
        cluster.sort(key=lambda x: (x[1], x[0]))  # Sort by y-coordinate, then by x-coordinate
        sorted_data.append(cluster)
    return sorted_data

sorted_data = sort_and_align_blocks(clustered_data)
```

```python
import pandas as pd

def structure_table_data(sorted_data):
    table = []
    for row in sorted_data:
        table_row = [cell[4] for cell in row]  # Extract the text from each cell
        table.append(table_row)
    return table

table_data = structure_table_data(sorted_data)
df = pd.DataFrame(table_data)
df
```
